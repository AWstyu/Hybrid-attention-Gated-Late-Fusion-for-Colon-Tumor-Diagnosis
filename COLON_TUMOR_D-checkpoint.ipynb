```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math

def process_all_tumor_masks(mask_path, min_area_threshold=50):
    mask = cv2.imread(mask_path, 0)
    _, binary_mask = cv2.threshold(mask, 127, 255, cv2.THRESH_BINARY)

    contours, hierarchy = cv2.findContours(binary_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    all_tumor_features = []
    mask_with_boxes = cv2.cvtColor(binary_mask, cv2.COLOR_GRAY2BGR)

    if not contours:
        print("No tumor-like regions detected in the mask.")
        return [], mask_with_boxes

    for i, contour in enumerate(contours):
        if cv2.contourArea(contour) < min_area_threshold:
            continue

        x, y, w, h = cv2.boundingRect(contour)
        bounding_box = (x, y, w, h)

        area = cv2.contourArea(contour)
        perimeter = cv2.arcLength(contour, True)
        
        # Min Feret Diameter (approximated from minAreaRect)
        rect = cv2.minAreaRect(contour)
        min_feret_width, min_feret_height = rect[1]
        min_feret_diameter = min(min_feret_width, min_feret_height)
        if min_feret_diameter <= 0: # Ensure non-zero for ratio calculation
            min_feret_diameter = 1e-6 # Small positive value to avoid division by zero
        
        # Max Feret Diameter (calculated from convex hull or contour points)
        max_feret_diameter = 0
        if len(contour) >= 3: # Need at least 3 points for convex hull
            convex_hull = cv2.convexHull(contour)
            if len(convex_hull) >= 2:
                hull_points = convex_hull.squeeze()
                for p_idx1 in range(len(hull_points)):
                    for p_idx2 in range(p_idx1 + 1, len(hull_points)):
                        p1 = hull_points[p_idx1]
                        p2 = hull_points[p_idx2]
                        dist = np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
                        if dist > max_feret_diameter:
                            max_feret_diameter = dist
            else: # Fallback if convex hull has less than 2 points
                contour_points = contour.squeeze()
                if len(contour_points) >= 2:
                    for p_idx1 in range(len(contour_points)):
                        for p_idx2 in range(p_idx1 + 1, len(contour_points)):
                            p1 = contour_points[p_idx1]
                            p2 = contour_points[p_idx2]
                            dist = np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
                            if dist > max_feret_diameter:
                                max_feret_diameter = dist
        elif len(contour) >= 2: # Fallback to original contour points if contour itself has points
             contour_points = contour.squeeze()
             for p_idx1 in range(len(contour_points)):
                for p_idx2 in range(p_idx1 + 1, len(contour_points)):
                    p1 = contour_points[p_idx1]
                    p2 = contour_points[p_idx2]
                    dist = np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
                    if dist > max_feret_diameter:
                        max_feret_diameter = dist
        
        # Ratio of Max and Min Feret Diameter
        feret_diameter_ratio = max_feret_diameter / min_feret_diameter

        euler_number = 1 

        eccentricity = 0
        if len(contour) >= 5:
            try:
                ellipse = cv2.fitEllipse(contour)
                major_axis = max(ellipse[1])
                minor_axis = min(ellipse[1])
                if major_axis > 0:
                    eccentricity = np.sqrt(1 - (minor_axis**2 / major_axis**2))
            except cv2.error:
                eccentricity = 0


        features = {
            "Bounding Box": bounding_box,
            "Area": area,
            "Perimeter": perimeter,
            "Min Feret Diameter": min_feret_diameter,
            "Max Feret Diameter": max_feret_diameter,
            "Feret Diameter Ratio": feret_diameter_ratio,
            "Euler Number": euler_number,
            "Eccentricity": eccentricity
        }
        all_tumor_features.append(features)

        cv2.rectangle(mask_with_boxes, (x, y), (x + w, y + h), (0, 255, 0), 2)

    return all_tumor_features, mask_with_boxes

def visualize_all_tumors(mask_path, min_area_threshold=50):
    all_tumor_features, mask_with_boxes = process_all_tumor_masks(mask_path, min_area_threshold)

    plt.figure(figsize=(10, 10))
    plt.imshow(mask_with_boxes, cmap='gray')
    plt.title(f"All Detected Tumor-like Regions ({len(all_tumor_features)} found)")
    plt.axis("off")
    plt.show()

    if not all_tumor_features:
        print("No tumor-like regions to display features for.")
        return

    print("Extracted Features for Each Tumor-like Region:")
    for idx, features in enumerate(all_tumor_features):
        print(f"\n--- Tumor Region {idx + 1} ---")
        print(f"  Bounding Box: {features['Bounding Box']}")
        print(f"  Area: {features['Area']:.2f}")
        print(f"  Perimeter: {features['Perimeter']:.2f}")
        print(f"  Min Feret Diameter: {features['Min Feret Diameter']:.2f}")
        print(f"  Max Feret Diameter: {features['Max Feret Diameter']:.2f}")
        print(f"  Feret Diameter Ratio: {features['Feret Diameter Ratio']:.2f}")
        print(f"  Euler Number: {int(features['Euler Number'])}")
        print(f"  Eccentricity: {features['Eccentricity']:.4f}")

mask_path = "C:\\COLON_CANCER DATASET\\PolypDB\\PolypDB_modality_wise\\COMBINED_PICS_MASK\\BLI_06_mask.jpg"
visualize_all_tumors(mask_path)
```

```python
import cv2
import numpy as np
import os
import math
import pandas as pd

def process_image_for_all_tumor_features(mask_path):
    original_mask = cv2.imread(mask_path, 0)
    if original_mask is None:
        print(f"Warning: Could not load mask image {os.path.basename(mask_path)}. Skipping.")
        return []

    _, binary_mask = cv2.threshold(original_mask, 127, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(binary_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    individual_tumor_data = []

    for contour in contours:
        area = float(cv2.contourArea(contour))
        perimeter = float(cv2.arcLength(contour, True))

        min_feret_diameter = 0.0
        max_feret_diameter = 0.0
        feret_diameter_ratio = 0.0

        if len(contour) >= 5:
            rect = cv2.minAreaRect(contour)
            min_feret_width, min_feret_height = rect[1]
            min_feret_diameter = float(min(min_feret_width, min_feret_height))
            
            if min_feret_diameter <= 0:
                min_feret_diameter = 1e-6

            if len(contour) >= 3:
                convex_hull = cv2.convexHull(contour)
                if len(convex_hull) >= 2:
                    hull_points = convex_hull.squeeze()
                    for p_idx1 in range(len(hull_points)):
                        for p_idx2 in range(p_idx1 + 1, len(hull_points)):
                            p1 = hull_points[p_idx1]
                            p2 = hull_points[p_idx2]
                            dist = math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
                            if dist > max_feret_diameter:
                                max_feret_diameter = float(dist)
                elif len(contour) >= 2:
                    contour_points = contour.squeeze()
                    for p_idx1 in range(len(contour_points)):
                        for p_idx2 in range(p_idx1 + 1, len(contour_points)):
                            p1 = contour_points[p_idx1]
                            p2 = contour_points[p_idx2]
                            dist = math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
                            if dist > max_feret_diameter:
                                max_feret_diameter = float(dist)
            elif len(contour) >= 2:
                contour_points = contour.squeeze()
                for p_idx1 in range(len(contour_points)):
                    for p_idx2 in range(p_idx1 + 1, len(contour_points)):
                        p1 = contour_points[p_idx1]
                        p2 = contour_points[p_idx2]
                        dist = math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
                        if dist > max_feret_diameter:
                            max_feret_diameter = float(dist)
        
            feret_diameter_ratio = max_feret_diameter / min_feret_diameter if min_feret_diameter > 0 else 0.0

        euler_number = 1.0

        eccentricity = 0.0
        if len(contour) >= 5:
            try:
                ellipse = cv2.fitEllipse(contour)
                major_axis = max(ellipse[1])
                minor_axis = min(ellipse[1])
                if major_axis > 0:
                    eccentricity = float(math.sqrt(1 - (minor_axis**2 / major_axis**2)))
            except cv2.error:
                eccentricity = 0.0

        individual_tumor_data.append({
            'area': area,
            'perimeter': perimeter,
            'min_feret_diameter': min_feret_diameter,
            'max_feret_diameter': max_feret_diameter,
            'feret_diameter_ratio': feret_diameter_ratio,
            'euler_number': euler_number,
            'eccentricity': eccentricity
        })
    
    return individual_tumor_data

def get_triplet(values):
    if not values:
        return (0.0, 0.0, 0.0)
    values_arr = np.array(values, dtype=float)
    if len(values) == 1:
        val = float(values_arr[0])
        return (val, val, val)
    else:
        return (float(np.min(values_arr)), float(np.mean(values_arr)), float(np.max(values_arr)))

def aggregate_shape_features_to_triplets(individual_tumor_data):
    if not individual_tumor_data:
        return {
            'area_triplet': (0.0, 0.0, 0.0),
            'perimeter_triplet': (0.0, 0.0, 0.0),
            'min_feret_diameter_triplet': (0.0, 0.0, 0.0),
            'max_feret_diameter_triplet': (0.0, 0.0, 0.0),
            'feret_diameter_ratio_triplet': (0.0, 0.0, 0.0),
            'euler_number_triplet': (0.0, 0.0, 0.0),
            'eccentricity_triplet': (0.0, 0.0, 0.0)
        }

    all_areas = []
    all_perimeters = []
    all_min_ferets = []
    all_max_ferets = []
    all_feret_ratios = []
    all_euler_numbers = []
    all_eccentricities = []

    for tumor in individual_tumor_data:
        all_areas.append(tumor['area'])
        all_perimeters.append(tumor['perimeter'])
        all_min_ferets.append(tumor['min_feret_diameter'])
        all_max_ferets.append(tumor['max_feret_diameter'])
        all_feret_ratios.append(tumor['feret_diameter_ratio'])
        all_euler_numbers.append(tumor['euler_number'])
        all_eccentricities.append(tumor['eccentricity'])

    aggregated_features = {
        'area_triplet': get_triplet(all_areas),
        'perimeter_triplet': get_triplet(all_perimeters),
        'min_feret_diameter_triplet': get_triplet(all_min_ferets),
        'max_feret_diameter_triplet': get_triplet(all_max_ferets),
        'feret_diameter_ratio_triplet': get_triplet(all_feret_ratios),
        'euler_number_triplet': get_triplet(all_euler_numbers),
        'eccentricity_triplet': get_triplet(all_eccentricities)
    }
    return aggregated_features

def process_directory_and_export_shape_triplets(directory_path, output_excel_path):
    image_shape_triplets = []
    
    if not os.path.isdir(directory_path):
        print(f"Error: Directory not found at {directory_path}")
        return

    sorted_filenames = sorted(os.listdir(directory_path))
    for filename in sorted_filenames:
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
            mask_path = os.path.join(directory_path, filename)
            
            individual_tumors_in_image = process_image_for_all_tumor_features(mask_path)
            
            aggregated_shape_triplets = aggregate_shape_features_to_triplets(individual_tumors_in_image)
            
            row_data = {'Image Name': filename}
            for feature_name, triplet_value in aggregated_shape_triplets.items():
                row_data[f'{feature_name}_min'] = triplet_value[0]
                row_data[f'{feature_name}_mean'] = triplet_value[1]
                row_data[f'{feature_name}_max'] = triplet_value[2]
            image_shape_triplets.append(row_data)

    if not image_shape_triplets:
        print("No valid tumor data found to export.")
        return

    df = pd.DataFrame(image_shape_triplets)
    df.to_excel(output_excel_path, index=False)
    print(f"Shape feature triplets exported to: {output_excel_path}")


image_directory = "C:\\COLON_CANCER DATASET\\PolypDB\\PolypDB_modality_wise\\WLI\\masks"
output_excel_file = "shape_feature_triplets_raw_WLI_mask.xlsx"

process_directory_and_export_shape_triplets(image_directory, output_excel_file)
```
